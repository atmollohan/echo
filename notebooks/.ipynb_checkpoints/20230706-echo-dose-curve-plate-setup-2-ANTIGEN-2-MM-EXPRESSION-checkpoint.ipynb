```python
import os
import numpy as np
from tqdm import tqdm #progress bar
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#make a list of wells in a 384-well plate
plate_columns=list(range(25))[1:]
plate_rows=["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L", "M", "N", "O", "P"]
wells=[]
for r in plate_rows:
    for c in plate_columns:
        wells.append(r+str(c))
        
wells96 = []
for r in plate_rows[0:8]:
    for c in plate_columns[0:12]:
        wells96.append(r+str(c))
```

```python
"""Note: units for ALL volumes is nL"""

#lets get sample information about the experiment you are running


doses = 5 #number of points in dose curve
doses2 = 4 #number points in "second" antigen dose curve
highest_dose = 40 #uM
vol_cellextract = 3425 #volume of cell extract in nL
vol_antigen = 675 #volume of antigen in nL
variants = ["nCov19-gblock", "NS1-gblock", "NS2-gblock", "NS3-gblock", "nCov19-plasmid", "NS1-mcrimson-plasmid"] #list of variants being tested
num_samples = len(variants)
dna_vol = 2000 #amnt DNA stock we have to add to source plate
dna_vol_per_well = 500 #amnt DNA to add to each well of dest plate

print(variants)
```

```python
#lets make a dataframe that functions as a plate map

def make_plate():
    plate_columns=list(range(25))[1:]
    plate_rows=["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L", "M", "N", "O", "P"]
    plate_df = pd.DataFrame(index=plate_rows, columns=plate_columns)
    return(plate_df)
```

```python
"""siddharth says to keep everything in terms of lists of 384 length, make lists of NaN's and add in values, then in
the back end it 
"""

source_sample_names = ['NaN']*384
source_sample_vol = [0]*384
dest_extract1_vol = [0]*384
dest_extract2_vol = [0]*384
dest_dna_vol = [0]*384
dest_antigen_vol = [0]*384
dest_antigen2_vol = [0]*384
dest_pbs_vol = [0]*384
dest_sample_names = ['NaN']*384

#lets start with adding the DNA samples to the source plate
i=0
for v in variants:
    source_sample_names[i]=v
    source_sample_vol[i]=dna_vol
    i+=2
    
#now let's calculate how much MM we will need, then add it to the source plate
master_mix = num_samples*(doses+doses2)*vol_cellextract
wells_mm_needed = (master_mix//40000) + 1
total_mm = (wells_mm_needed-1)*55000+(master_mix%40000)
i=192


while wells_mm_needed>1:
    source_sample_names[i]='MM+enh'
    source_sample_vol[i]=55000
    i+=1
    wells_mm_needed-=1

source_sample_names[i]='MM+enh'
source_sample_vol[i]=15000+(master_mix%40000)
i+=1



#add second master mix to the plate
wells_mm_needed = (master_mix//40000) + 1

while wells_mm_needed>1:
    source_sample_names[i]='MM-enh'
    source_sample_vol[i]=55000
    i+=1
    wells_mm_needed-=1

source_sample_names[i]='MM-enh'
source_sample_vol[i]=15000+(master_mix%40000)
i+=1

print("amount of each master mix to make: " + str(total_mm/1000) + "uL")
#print(source_sample_vol)
```

```python
#now let's add the antigen & PBS wells
#we'll first need to calculate how much we'll need for each


amnt_antigen_to_add = [675, 225, 75, 25, 0] #approx 1:3 dilutions, in nL
amnt_pbs_to_add = []
for a in amnt_antigen_to_add:
    amnt_pbs_to_add.append(vol_antigen-a)
    
#adding antigen to source plate
total_antigen_vol = 2*sum(amnt_antigen_to_add)*num_samples
wells_antigen_needed = total_antigen_vol//40000 + 1

while wells_antigen_needed>1:
    source_sample_names[i]='antigen1'
    source_sample_vol[i]=55000
    i+=1
    wells_antigen_needed-=1

source_sample_names[i]='antigen1'
source_sample_vol[i]=15000+(total_antigen_vol%40000)
i+=1


#adding 2nd antigen to source plate
total_antigen2_vol = 2*sum(amnt_antigen_to_add[0:doses2])*num_samples
wells_antigen2_needed = total_antigen2_vol//40000 + 1

while wells_antigen2_needed>1:
    source_sample_names[i]='antigen2'
    source_sample_vol[i]=55000
    i+=1
    wells_antigen2_needed-=1

source_sample_names[i]='antigen2'
source_sample_vol[i]=15000+(total_antigen2_vol%40000)
i+=1



#adding pbs to source plate
total_pbs_vol = (sum(amnt_pbs_to_add)*num_samples) + (sum(amnt_pbs_to_add[0:doses2])*num_samples)
wells_pbs_needed = total_pbs_vol//40000 + 1

while wells_pbs_needed>1:
    source_sample_names[i]='buffer'
    source_sample_vol[i]=55000
    i+=1
    wells_pbs_needed-=1

source_sample_names[i]='buffer'
source_sample_vol[i]=15000+(total_pbs_vol%40000)
i+=1

print(source_sample_names)
print(source_sample_vol)

```

```python
place = 0

#for 1st master mix
for v in variants:
    #1st antigen
    for d in range(doses):
        dest_extract1_vol[place] = vol_cellextract
        dest_dna_vol[place] = dna_vol_per_well
        dest_antigen_vol[place] = amnt_antigen_to_add[d]
        dest_pbs_vol[place] = amnt_pbs_to_add[d]
        dest_sample_names[place] = v + "_dose" + str(d) + "_+enh"
        place+=1
    #2nd antigen
    for d in range(doses2):
        dest_extract1_vol[place] = vol_cellextract
        dest_dna_vol[place] = dna_vol_per_well
        dest_antigen2_vol[place] = amnt_antigen_to_add[d]
        dest_pbs_vol[place] = amnt_pbs_to_add[d]
        dest_sample_names[place] = v + "_dose" + str(d+doses) + "_+enh"
        place+=1
        
#for 2nd master mix
for v in variants:
    #1st antigen
    for d in range(doses):
        dest_extract2_vol[place] = vol_cellextract #adjust this for dna input
        dest_dna_vol[place] = dna_vol_per_well
        dest_antigen_vol[place] = amnt_antigen_to_add[d]
        dest_pbs_vol[place] = amnt_pbs_to_add[d]
        dest_sample_names[place] = v + "_dose" + str(d) + "_-enh"
        place+=1
    #2nd antigen
    for d in range(doses2):
        dest_extract2_vol[place] = vol_cellextract #adjust this for dna input
        dest_dna_vol[place] = dna_vol_per_well
        dest_antigen2_vol[place] = amnt_antigen_to_add[0:doses2][d]
        dest_pbs_vol[place] = amnt_pbs_to_add[0:doses2][d]
        dest_sample_names[place] = v + "_dose" + str(d+doses) + "_-enh"
        place+=1

        
print(dest_sample_names)
print(dest_antigen_vol)
print(dest_antigen2_vol)
```

```python
echo_protocol = pd.DataFrame(columns=["Sample Name", "Source Plate Name", "Source Well", "Destination Well", 
                                      "Transfer Volume", "Destination Plate Name", "Source Plate Type"])

def add_line(sample, source, dest, vol, df):
    new = pd.Series([sample,'af_src001',source,dest,vol,"dest_plate","384PP_AQ_BP"], index=df.columns)
    return(df.append(new, ignore_index=True))


for well in range(384):
    #lets first add cell extract to the protocol
    if dest_extract1_vol[well] > 0:
        #first need to find the well in the source plate
        for x in range(192,384):
            if source_sample_names[x]=='MM+enh' and source_sample_vol[x]>15000+dest_extract1_vol[well]:
                source_well = wells[x]
                source_sample_vol[x] = source_sample_vol[x]-dest_extract1_vol[well]
                break          
        echo_protocol = add_line("cell_extract+enh", source_well, wells[well], dest_extract1_vol[well], echo_protocol)
    elif dest_extract2_vol[well] > 0:
        #first need to find the well in the source plate
        for x in range(192,384):
            if source_sample_names[x]=='MM-enh' and source_sample_vol[x]>15000+dest_extract2_vol[well]:
                source_well = wells[x]
                source_sample_vol[x] = source_sample_vol[x]-dest_extract2_vol[well]
                break          
        echo_protocol = add_line("cell_extract-end", source_well, wells[well], dest_extract2_vol[well], echo_protocol)

#now lets add the dna to the protocol
for well in range(384):
    if dest_dna_vol[well] > 0:
        sample_dna = dest_sample_names[well].split('_')[0]    
        for x in range(192):
            if source_sample_names[x]==sample_dna:
                source_well = wells[x]
                break
        echo_protocol = add_line(sample_dna, source_well, wells[well], dest_dna_vol[well], echo_protocol)        
        
#next let's add the pbs to the protocol
for well in range(384):
    if dest_pbs_vol[well] > 0:
        #first need to find the well in the source plate
        for x in range(192,384):
            if source_sample_names[x]=='buffer' and source_sample_vol[x]>15000+dest_pbs_vol[well]:
                source_well = wells[x]
                source_sample_vol[x] = source_sample_vol[x]-dest_pbs_vol[well]
                break          
        echo_protocol = add_line("buffer", source_well, wells[well], dest_pbs_vol[well], echo_protocol)

#lastly add the antigen
for well in range(384):
    if dest_antigen_vol[well] > 0: #1st antigen
        #first need to find the well in the source plate
        for x in range(192,384):
            if source_sample_names[x]=='antigen1' and source_sample_vol[x]>15000+dest_antigen_vol[well]:
                source_well = wells[x]
                source_sample_vol[x] = source_sample_vol[x]-dest_antigen_vol[well]
                break          
        echo_protocol = add_line("antigen1", source_well, wells[well], dest_antigen_vol[well], echo_protocol)
    elif dest_antigen2_vol[well] > 0: #2nd antigen
        #first need to find the well in the source plate
        for x in range(192,384):
            if source_sample_names[x]=='antigen2' and source_sample_vol[x]>15000+dest_antigen2_vol[well]:
                source_well = wells[x]
                source_sample_vol[x] = source_sample_vol[x]-dest_antigen2_vol[well]
                break          
        echo_protocol = add_line("antigen2", source_well, wells[well], dest_antigen2_vol[well], echo_protocol)




print(echo_protocol)

#export protocol to csv
echo_protocol.to_csv('20230719_echo_protocol.csv')   

```
